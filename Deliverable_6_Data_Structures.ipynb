```python
import pandas as pd
```

```python
def Odd(n):
  Odd_list = []
  for nth in range(0,n):
    Odd = nth*2+1
    Odd_list.append(Odd)
  return Odd_list
```

```python
def Even(n):
  Even_list = []
  for nth in range(0,n):
    Even = nth*2
    Even_list.append(Even)
  return Even_list
```

```python
def fibonacci(n):
  fibonacci=[0,1]
  if (n == 1):
    fibonacci.pop()
  for nth in range(0,n-2):
    val = fibonacci[-1] + fibonacci[-2]
    fibonacci.append(val)
  return fibonacci
```

```python
try:
  length = int(input("Please enter an integer value greater than 0: "))
except ValueError:
  print("ERROR: Couldn't convert data to an integer.")
if (length <= 0):
  print("ERROR: Integer must be greater than 0")
else:
  Odd_list = Odd(length)
  Even_list = Even(length)
  Fibonacci_list = fibonacci(length)

  df = pd.DataFrame({'Even': Even_list, 'Odd': Odd_list, 'Fibonacci': Fibonacci_list})

df
```
